# Requesting drug page

```python
url_drug = "https://www.mims.com" + A_list[14].get("href")
page_drug = requests.get(url_drug)
soup_drug = BeautifulSoup(page_drug.content, 'html.parser')
```

```python
drug_name = soup_drug.find("div", class_="div-brand").get_text().strip()
drug_name
```

```python
drug_generic_html = soup_drug.find("h2")
if drug_generic_html is None:
    drug_generic = "NULL"
else:
    drug_generic = drug_generic_html.get_text().strip()
drug_generic
```

```python
drug_section = soup_drug.find_all("div", class_= ["monograph-section-header","monograph-section-content"] )
drug_section
```

```python
for index, ele in enumerate(drug_section):
    if "MIMS Class" in ele.get_text():
        MIMS_class = drug_section[index + 1].get_text().strip()
    if "ATC" in ele.get_text():
        ATC_class = drug_section[index + 1].get_text().strip()
    if "Regulatory" in ele.get_text():
        Regulatory_class = drug_section[index + 1].get_text().strip()
```
