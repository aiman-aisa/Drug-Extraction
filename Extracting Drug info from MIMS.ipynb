# Create dictionary

```python
column_names = ["Drug_Id", 
                "Drug_Name", 
                "Type", 
                "Generic_Drug", 
                "MIMS_Class", 
                "ATC_Classification", 
                "Regulatory_Classification",
                "Remarks"]
drug_dict= dict.fromkeys(column_names)
drug_dict['Drug_Id'] = []
drug_dict['Drug_Name'] = []
drug_dict['Type'] = []
drug_dict['Generic_Drug'] = []
drug_dict['MIMS_Class'] = []
drug_dict['ATC_Classification'] = []
drug_dict['Regulatory_Classification'] = []
drug_dict['Remarks'] = []

drug_dict
```

```python
drug_Id = 0
drug_type = "BRAND"
for index, drug in enumerate(A_list):
    #print(str(index) + " " + drug.get("href"))
    # get url for drug page
    url_drug = "https://www.mims.com" + drug.get("href")
    page_drug = requests.get(url_drug)
    
    # get content of the page
    soup_drug = BeautifulSoup(page_drug.content, 'html.parser')
    
    # define drug id
    drug_Id = drug_Id + 1
    
    # define drug name
    drug_name = soup_drug.find("div", class_="div-brand").get_text().strip()
    
    # get drug generic if exist
    drug_generic_html = soup_drug.find("h2")
    if drug_generic_html is None:
        drug_generic = "NULL"
    else:
        drug_generic = drug_generic_html.get_text().strip()
    
    # find the drug section list
    drug_section = soup_drug.find_all("div", class_= ["monograph-section-header","monograph-section-content"])
    
    # initializing the object for columns in drug section
    MIMS_class = "NULL"
    ATC_class = "NULL"
    Regulatory_class = "NULL"
    
    # iterating through the drug section list
    for index, ele in enumerate(drug_section):
        if "MIMS Class" in ele.get_text():
            MIMS_class = drug_section[index + 1].get_text().strip()
        if "ATC" in ele.get_text():
            ATC_class = drug_section[index + 1].get_text().strip()
        if "Regulatory" in ele.get_text():
            Regulatory_class = drug_section[index + 1].get_text().strip()
    
    # get the drug remark if exists
    drug_remark_html = soup_drug.find("div", class_="discontinued")
    if drug_remark_html is None:
        drug_remark = "NULL"
    else:
        drug_remark = drug_remark_html.get_text().strip()
        
    # updating the object into drug_dict
    drug_dict['Drug_Id'].append(drug_Id)
    drug_dict['Drug_Name'].append(drug_name)
    drug_dict['Type'].append(drug_type)
    drug_dict['Generic_Drug'].append(drug_generic)
    drug_dict['MIMS_Class'].append(MIMS_class)
    drug_dict['ATC_Classification'].append(ATC_class)
    drug_dict['Regulatory_Classification'].append(Regulatory_class)
    drug_dict['Remarks'].append(drug_remark)
    
```

```python
df = pd.DataFrame.from_dict(drug_dict)
```
